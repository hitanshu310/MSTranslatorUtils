```python
import pandas as pd
import numpy as np
```

```python
## visualizing nllb

## There are different models of nllb with different parameter values

## MoE -- facebook/nllb-moe-54b
## facebook/nllb-200-3.3B
## facebook/nllb-200-distilled-600M
```

```python
nllb_54b_metrics = pd.read_csv('metrics_nllb_54b_metrics.csv')
nllb_54b_metrics
```

```python
nllb_54b_metrics[['from_lang', 'to_lang']] = nllb_54b_metrics['direction'].str.split('-',  expand=True)
```

```python
nllb_54b_metrics
```

```python
bleu_200 = nllb_54b_metrics[['from_lang', 'to_lang', 'spBLEU(spm-200)']]
```

```python
bleu_200['spBLEU(spm-200)'] = np.where(bleu_200['spBLEU(spm-200)'] < 30, 0, bleu_200['spBLEU(spm-200)'])
```

```python
bleu_200 = bleu_200.pivot(index='from_lang', columns='to_lang', values='spBLEU(spm-200)')
```

```python
import matplotlib.pyplot as plt
import matplotlib as ml

bleu_200 = bleu_200.fillna(0)
vmax = max(bleu_200.values.reshape(202 * 202,))
norm = ml.colors.Normalize(0, vmax=vmax)

fig, ax = plt.subplots(figsize=(101,101))
heatmap = plt.pcolor(bleu_200, cmap='Greens', norm=norm)

ax.set_xticks(np.arange(bleu_200.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(bleu_200.shape[0]) + 0.5, minor=False)

plt.xticks(rotation=90)

ax.set_xticklabels(bleu_200.index)
ax.set_yticklabels(bleu_200.columns)

plt.show()
```
